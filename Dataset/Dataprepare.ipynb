```python
def prep(inputfile, outputfile):
    # Open the file with 'utf-8' encoding
    with open(inputfile, 'r', encoding='utf-8') as f:
        # Load JSON data from file
        data = json.load(f)
    with open(outputfile , 'w', encoding='utf-8') as f:
        for i in range(len(data)): # Loop through the data (each row is a JSON object)
            for j in range(len(data[i]['tokens'])): # Loop through the tokens in each row i.e. the 'tokens' key
                f.write(data[i]['tokens'][j]['token'] + ' ')
                for k in range(len(data[i]['tokens'][j]['tags'])): # Loop through the tags in each row i.e. the 'tags' key
                    f.write(data[i]['tokens'][j]['tags'][k]['value'] + ' ')
                    if (data[i]['tokens'][j]['tags'][k]['tags']) is not None: # Check if there are any subtags
                        f.write(data[i]['tokens'][j]['tags'][k]['tags'][0]['value'] + '\n')
                    else: f.write('\n') # If there are no subtags, write a newline character
            f.write('\n') # Write a newline character at the end of the row
```

```python
import csv
import json

def prepcsv(inputfile, outputfile):
    with open(inputfile, 'r', encoding='utf-8') as f:
        data = json.load(f)

    with open(outputfile, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(["Token", "Tags", "Subtags"])  # Write the header row

        for i in range(len(data)):
            for j in range(len(data[i]['tokens'])):
                token = data[i]['tokens'][j]['token']
                tags = data[i]['tokens'][j]['tags']

                for k in range(len(tags)):
                    tag = tags[k]['value']
                    subtags = tags[k].get('tags')

                    if subtags is not None:
                        subtag = subtags[0]['value']
                    else:
                        subtag = ''

                    writer.writerow([token, tag, subtag])
```

```python
# libraries
import string
import codecs
import numpy as np
import pandas as pd
import re
import io
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import arabicstopwords.arabicstopwords as stp

stop_words = set(stp.stopwords_list())


def remove_punctuation(value):
    result = ""
    for c in value:
        # If char is not punctuation, add it to the result.
        if c not in string.punctuation:
            result += c
   # value = result
    value = result

    return value
# -----------------------------------------------------------#


def remove_questionmark(value):
    value = re.sub("؟", "", value)
    value = re.sub("?", "", value)
    value = re.sub(",", "", value)
    value = re.sub(".", "", value)
    value = re.sub("،", "", value)
    value = re.sub("[إأٱآاٲ]", "ا", value)
    value = re.sub("ة", "ه", value)
  #  value=re.sub(r'[a-zA-Z?]', '', value).strip() # remove non arabic charachter

    value = re.sub(r" ?\([^)]+\)", "", value)  # remove parnthesses

    return (value)

# -----------------------------------------------------------#


def deletestopwords(body):

    delstopwords = ""  # this is string to be used for correct words initial is empty

    spacesplit = body.split()
    # counter = len(spacesplit)
    # print counter
    body = ""

    for r in spacesplit:
        if not r in stop_words:

            # add value not within stop words to this string
            delstopwords = delstopwords + " " + r
            # print delstopwords
   # print delstopwords
    body = delstopwords
  #  print (body)
    return body
# --------------------------------------#


in_file = codecs.open(
    r"C:\Users\shawn\Desktop\Active work\Arabic Ner\JSON-FLAT\train.csv", 'r', encoding="utf_8")
out_file = codecs.open("punc_pre.csv",
                       'w', buffering=0, encoding="utf_8")


reviews = in_file.readlines()
# --------------------------------------------------#


for i in range(0, len(reviews)):

    parts = [4]
    Text2 = reviews[i]
    parts = Text2.split()

    Question = parts[0]

 #   Question = remove_punctuation(Question)
    Question = remove_questionmark(Question)
    Question = deletestopwords(Question)

    line = u"%s\n" % (Question)

    out_file.write(line)
```
